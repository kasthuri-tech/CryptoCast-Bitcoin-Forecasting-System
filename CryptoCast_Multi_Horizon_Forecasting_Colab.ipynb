## Phase 1: Data Cleaning & Number Standardization Functions
**Note:** Spreadsheets downloaded from the web store currency and volume numbers with extra text symbols like commas, percentage signs, and letters. This section sets up helper functions to strip away those unwanted text symbols so Python can perform mathematical calculations on them.

### What Each Step Does:
1. **Step 1 (Price Cleaning):** Removes commas from raw dollar text strings (example: `"67,211.9"` becomes `67211.9`).
2. **Step 2 (Volume Scaling):** Converts letter shortcuts into actual numbers (`K` becomes thousands, `M` becomes millions, `B` becomes billions).
3. **Step 3 (Percentage Trimming):** Removes the trailing percent symbol from daily momentum numbers so they can be treated as standard numbers.

```python
import os
import sqlite3
import pandas as pd
import numpy as np

def remove_commas_from_price_string(raw_text):
    if pd.isna(raw_text) or raw_text == "" or raw_text == "-": return np.nan
    cleaned_text = str(raw_text).replace(",", "").strip()
    try: return float(cleaned_text)
    except ValueError: return np.nan

def convert_volume_letters_to_numbers(volume_text):
    if pd.isna(volume_text) or volume_text == "" or volume_text == "-": return np.nan
    volume_text = str(volume_text).strip().upper()
    try:
        if volume_text.endswith('K'): return float(volume_text.replace('K', '')) * 1e3
        elif volume_text.endswith('M'): return float(volume_text.replace('M', '')) * 1e6
        elif volume_text.endswith('B'): return float(volume_text.replace('B', '')) * 1e9
        else: return float(volume_text)
    except ValueError: return np.nan

def remove_percentage_symbol(percentage_text):
    if pd.isna(percentage_text) or percentage_text == "" or percentage_text == "-": return np.nan
    percentage_text = str(percentage_text).replace("%", "").strip()
    try: return float(percentage_text)
    except ValueError: return np.nan

print("[+] Data cleaning utilities active.")
```
